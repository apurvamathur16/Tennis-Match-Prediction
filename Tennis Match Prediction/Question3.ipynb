```python
def normalize(dff,col_name_list):
    result = dff.copy()
    for feature_name in col_name_list:
        max_value = dff[feature_name].max()
        min_value = dff[feature_name].min()
        result[feature_name] = (dff[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
df = normalize(df,num_cols)
df = pd.get_dummies(df, columns=cat_cols) 
```

```python
def getPlayerNames():
    player_names = {}
    for index, row in df.iterrows():
        if(row['Player1_name'] not in player_names.keys()):
            player_names[row['Player1_name']] = 1
        else:
            player_names[row['Player1_name']] = player_names[row['Player1_name']] + 1
        
        if(row['Player2_name'] not in player_names.keys()):
            player_names[row['Player2_name']] = 1
        else:
            player_names[row['Player2_name']] = player_names[row['Player2_name']] + 1
    return player_names
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
def logisticRegression(c,X_train, X_test, y_train, y_test):
    clf = LogisticRegression(penalty = 'l2', C = c,random_state = 0,solver='sag')
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    test_acc = metrics.accuracy_score(y_test, y_pred)*100
    mse = metrics.mean_squared_error(y_test,y_pred)
    r2s = metrics.r2_score(y_test,y_pred)
    training_acc = clf.score(X_train, y_train)*100
    h = []
    h.append(test_acc)
    h.append(training_acc)
    h.append(mse)
    h.append(r2s)
    h.append(y_pred)
    return h
    
```

```python
from sklearn import svm
from sklearn.metrics import classification_report, confusion_matrix 

def SVM(X_train, X_test, y_train, y_test):
    clf = svm.SVC(kernel="rbf", C=0.025, probability=True)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    test_acc = metrics.accuracy_score(y_test, y_pred)*100
    mse = metrics.mean_squared_error(y_test,y_pred)
    r2s = metrics.r2_score(y_test,y_pred)
    training_acc = clf.score(X_train, y_train)*100
    h = []
    h.append(test_acc)
    h.append(training_acc)
    h.append(mse)
    h.append(r2s)
    h.append(y_pred)
    return h
```

```python
def randomForest(est,dep,X_train, X_test, y_train, y_test):
    rdforest = RandomForestClassifier(n_estimators=est,max_depth=dep)
    rdforest.fit(X_train, y_train)
    rdpreds = rdforest.predict(X_test)
    test_acc = metrics.accuracy_score(y_test, rdpreds)*100
    mse = metrics.mean_squared_error(y_test,rdpreds)
    r2s = metrics.r2_score(y_test,rdpreds)
    training_acc = rdforest.score(X_train, y_train)*100
    h = []
    h.append(test_acc)
    h.append(training_acc)
    h.append(mse)
    h.append(r2s)
    h.append(rdpreds)
    return h
```

```python
from sklearn.naive_bayes import GaussianNB

def naiveBayes(X_train, X_test, y_train, y_test):
    gnb = GaussianNB()
    y_pred = gnb.fit(X_train, y_train).predict(X_test)
    test_acc = metrics.accuracy_score(y_test, y_pred)*100
    mse = metrics.mean_squared_error(y_test,y_pred)
    r2s = metrics.r2_score(y_test,y_pred)
    training_acc = gnb.score(X_train, y_train)*100
    h = []
    h.append(test_acc)
    h.append(training_acc)
    h.append(mse)
    h.append(r2s)
    h.append(y_pred)
    return h
```

```python
import operator
def runModels(dff,player):
    Y = []
    for index2, row2 in dff.iterrows():
        if(row2['Player1_name']==player):
            Y.append(row2['player1_form'])
        else:
            Y.append(row2['player2_form'])
    
    dff['y'] = Y
    dff = dff.drop(['Player1_name'], axis=1)
    dff = dff.drop(['Player2_name'], axis=1)
    
    y = pd.DataFrame(dff['y'])
    
    dff = dff.drop(['y'], axis=1)
    X = dff
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    
    lr = logisticRegression(95,X_train, X_test, y_train, y_test)
    svm = SVM(X_train, X_test, y_train, y_test)
    rf = randomForest(150,10,X_train, X_test, y_train, y_test)
    nb = naiveBayes(X_train, X_test, y_train, y_test)
    
    dic = {}
    dic['LR'] = lr[0]
    dic['SVM'] = svm[0]
    dic['rf'] = rf[0]
    dic['nb'] = nb[0]
    
    bm = max(dic.items(), key=operator.itemgetter(1))[0]
    if(bm=='LR'):
        print("\t\tBest Model is Logistic Regression")
        print("\t\tTraining Accuracy: ",lr[1])
        print("\t\tTest Accuracy: ",lr[0])
        ff = {}
        for j in lr[4]:
            if(j in ff.keys()):
                ff[j] = ff[j]+1
            else:
                ff[j] = 1
        form = max(ff.items(), key=operator.itemgetter(1))[0]
        print("\t\tPredicted Form: ",form)
    elif(bm=='SVM'):
        print("\t\tBest Model is SVM with rbf Kernel")
        print("\t\tTraining Accuracy: ",svm[1])
        print("\t\tTest Accuracy: ",svm[0])
        ff = {}
        for j in svm[4]:
            if(j in ff.keys()):
                ff[j] = ff[j]+1
            else:
                ff[j] = 1
        form = max(ff.items(), key=operator.itemgetter(1))[0]
        print("\t\tPredicted Form: ",form)
    elif(bm=='rf'):
        print("\t\tBest Model is Random Forest")
        print("\t\tTraining Accuracy: ",rf[1])
        print("\t\tTest Accuracy: ",rf[0])
        ff = {}
        for j in rf[4]:
            if(j in ff.keys()):
                ff[j] = ff[j]+1
            else:
                ff[j] = 1
        form = max(ff.items(), key=operator.itemgetter(1))[0]
        print("\t\tPredicted Form: ",form)
    elif(bm=='nb'):
        print("\t\tBest Model is Logistic Regression")
        print("\t\tTraining Accuracy: ",nb[1])
        print("\t\tTest Accuracy: ",nb[0])
        ff = {}
        for j in nb[4]:
            if(j in ff.keys()):
                ff[j] = ff[j]+1
            else:
                ff[j] = 1
        form = max(ff.items(), key=operator.itemgetter(1))[0]
        print("\t\tPredicted Form: ",form)
```

```python
def predictForm():
    player_names = getPlayerNames()
    L  = list(df)
    #print(L)
    tot_players = len(player_names)
    k=1
    for i in player_names.keys():
        print("Remaining Players: ",tot_players-k)
        k = k + 1
        if(player_names[i]>=800):
            dff = pd.DataFrame(columns = L)
            for index, row in df.iterrows():
                if(row['Player1_name']==i or row['Player2_name']==i):
                    dff = dff.append(row)
        
            if(dff.shape[0]>=800):
                print("\tPlayer Name: ",i)
                print("\tdata: ",dff.shape)    
                runModels(dff,i)
```

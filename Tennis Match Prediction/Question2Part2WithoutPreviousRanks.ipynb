```python
import numpy as np
import pandas as pd
```

```python
PATH = '/data/'
ext = '.csv'
#full data from 1991
start1 = 2018 #1968
end1 = 2018
ft1 = 'atp_matches_'
```

```python
def readfiles(ft,st,en):
    l = []
    for i in range(st,en+1):
        p = ft + str(i)+ ext
        #print(p)
        df = pd.read_csv(p)
        l.append(df)
        #print(df.head(2))
    return l
```

```python
f1 = readfiles(ft1,start1,end1)
```

```python
all_matches = pd.concat(f1)
all_matches.describe()
```

```python
all_matches.info()
#all_matches=all_matches.dropna(axis=0,how='any')
all_matches = all_matches[np.isfinite(all_matches['loser_rank'])]
all_matches = all_matches[np.isfinite(all_matches['winner_rank'])]

#print("shape is ",all_matches.shape)
```

```python
to_drop = ['tourney_id','tourney_date','winner_id','winner_name','loser_id','loser_name','score']
cat_cols = ['tourney_name','surface','draw_size','tourney_level','winner_entry','winner_hand','winner_ioc',
            'loser_entry','loser_hand','loser_ioc','best_of','round']
num_cols = ['match_num','winner_seed','winner_ht','winner_age','winner_rank_points',
            'loser_seed','loser_ht','loser_age','loser_rank_points','minutes',
            'w_ace','w_df','w_svpt','w_1stIn','w_1stWon','w_2ndWon','w_SvGms','w_bpSaved',
             'w_bpFaced','l_ace','l_df','l_svpt','l_1stIn','l_1stWon','l_2ndWon','l_SvGms','l_bpSaved','l_bpFaced']
num_cols1 = ['match_num','winner_seed','winner_ht','winner_age',
            'loser_seed','loser_ht','loser_age','minutes',
            'w_ace','w_df','w_svpt','w_1stIn','w_1stWon','w_2ndWon','w_SvGms','w_bpSaved',
             'w_bpFaced','l_ace','l_df','l_svpt','l_1stIn','l_1stWon','l_2ndWon','l_SvGms','l_bpSaved','l_bpFaced']
all_matches = all_matches.drop(to_drop, axis=1)
```

```python
#all_matches = all_matches.fillna(all_matches.mean())
for i in cat_cols:
    all_matches[i] = all_matches[i].replace(np.NaN, all_matches[i].mode())

for i in num_cols:
    all_matches[i] = all_matches[i].replace(np.NaN,all_matches[i].mean())
    
all_matches = pd.get_dummies(all_matches, columns=cat_cols)



def normalize(dff,col_name_list):
    result = dff.copy()
    for feature_name in col_name_list:
        max_value = dff[feature_name].max()
        min_value = dff[feature_name].min()
        result[feature_name] = (dff[feature_name] - min_value) / (max_value - min_value)
    return result


all_matches = normalize(all_matches,num_cols)
```

```python
loserrank = pd.DataFrame(all_matches['loser_rank'])
winnerrank = pd.DataFrame(all_matches['winner_rank'])


all_matches = all_matches.drop(['winner_rank','loser_rank'], axis=1)

loserrankDF=all_matches.copy()
winnerrankDF=all_matches.copy()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
def randomForest(est,dep,X_train,X_test, y_train,y_test):
    print("Random Forest")
    rdforest = RandomForestClassifier(n_estimators=est,max_depth=dep)
    rdforest.fit(X_train, y_train)
    rdpreds = rdforest.predict(X_test)
    print ("\tMean Squared Error: ",metrics.mean_squared_error(y_test,rdpreds))
    print ("\tR2 Score: ",metrics.r2_score(y_test,rdpreds))
    print ("\tTraining Accuracy: ",rdforest.score(X_train, y_train)*100)
    print ("\tTest Accuracy: ",metrics.accuracy_score(np.round(y_test),np.round(rdpreds))*100)
randomForest(100,70,X1_train,X1_test, y1_train,y1_test)
#randomForest(100,70,X2_train,X2_test, y2_train,y2_test)
```
